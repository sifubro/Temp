```python
import requests
from bs4 import BeautifulSoup
```

```python
API_URL = {
    'google' : 'http://www.google.com/searchbyimage/upload',
    'bing' : '',
    'yandex' : '',
    'baidu' : ''
    
}

API_LIST = ['google', 'bing', 'yandex', 'baidu']

def ReverseImageSearchLabeller(imgPath, API='google'):
    '''
    inputs
    ------
    imgPath: image path that we would like to label.
    API : any on the following 'google', 'bing', ...
    
    output: label of the image (string)
    '''
    #imgPath = 'C://Users/30694/Desktop/Google_Image_Search/Pascal_Sample_occluded/bus1.jpg'
    
    if API not in API_URL:
        raise Exception('Invalid API name')
    
    searchUrl = API_URL[API]
    multipart = {'encoded_image': (imgPath, open(imgPath, 'rb')), 'image_content': ''}
    response = requests.post(searchUrl, files=multipart, allow_redirects=False)
    searchUrl = response.headers['Location']
    response = requests.get(searchUrl, headers={
         "user-agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/53.0.2785.143 Safari/537.36"})
    html = response.content
    soup = BeautifulSoup(html, "html.parser")
    
    
    # for google
    if API =='google':
        try:
            label = soup.findAll('h2', {'data-attrid':'title'})[0].span.text
        except:
            print("{} API couldn't find a label. Trying 'bing' API".format(API))
            ReverseImageSearchLabeller(imgPath, API='bing') 
    elif API =='bing':
        try:
            # !TODO! check this label = soup.findAll(...)
            label = soup.findAll('h2', {'data-attrid':'title'})[0].span.text
        except:
            print("{} API couldn't find a label. Trying 'yandex' API".format(API))
            ReverseImageSearchLabeller(imgPath, API='yandex') 
    elif API =='yandex':
        try:
            # !TODO! check this label = soup.findAll(...)
            label = soup.findAll('h2', {'data-attrid':'title'})[0].span.text
        except:
            print("{} API couldn't find a label. Trying 'baidu' API".format(API))
            ReverseImageSearchLabeller(imgPath, API='baidu') 
    elif API =='baidu':
        try:
            # !TODO! check this label = soup.findAll(...)
            label = soup.findAll('h2', {'data-attrid':'title'})[0].span.text
        except:
            print("{} API couldn't find a label. Trying 'google' API".format(API))
            ReverseImageSearchLabeller(imgPath, API='google') 
    
    
    return label
```
